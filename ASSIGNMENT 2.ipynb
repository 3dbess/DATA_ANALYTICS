```python
import pandas as pd


url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00519/heart_failure_clinical_records_dataset.csv"
df = pd.read_csv(url)


print(df.head())
```

```python
import matplotlib.pyplot as plt


gender_counts = df['sex'].value_counts()


plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['blue', 'pink'])
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python

plt.figure(figsize=(8, 6))
plt.plot(df['age'], df['serum_creatinine'], marker='o', linestyle='-')
plt.title('Age vs Serum Creatinine')
plt.xlabel('Age')
plt.ylabel('Serum Creatinine')
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(8, 6))
plt.scatter(df['age'], df['ejection_fraction'], color='green', alpha=0.5)
plt.title('Age vs Ejection Fraction')
plt.xlabel('Age')
plt.ylabel('Ejection Fraction')
plt.grid(True)
plt.show()
```

```python

smoking_counts = df['smoking'].value_counts()


plt.figure(figsize=(8, 6))
plt.pie(smoking_counts, labels=smoking_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Distribution of Smoking Status')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df['serum_creatinine'], vert=False)
plt.title('Distribution of Serum Creatinine')
plt.xlabel('Serum Creatinine')
plt.grid(True)
plt.show()
```
